### This file explains and tests phase kickback effect

```python
from qiskit.visualization import plot_bloch_vector, plot_bloch_multivector
from qiskit import QuantumCircuit, assemble, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_bloch_vector
from math import sqrt, pi

sim = AerSimulator()
```

```python
# example of how to plot the state of a single qubit

qc = QuantumCircuit(1)
qc.x(0)
qc.save_statevector()
job = sim.run(transpile(qc, sim), shots=256)
state = job.result().get_statevector()
plot_bloch_multivector(state)
```

```python
# Phase Kickback Effect
phi = pi/4
qc = QuantumCircuit(4)
qc.h(0)
qc.x(1)
qc.cp(phi, 0, 1)
qc.h(2)
qc.x(3)
qc.p(phi, 3)
qc.save_statevector()
qc.draw(output='mpl')
```

```python
job = sim.run(transpile(qc, sim), shots=256)
state = job.result().get_statevector()
plot_bloch_multivector(state)
```

```python
# Phase Kickback Effect
phi = pi/4
disturb = pi/8
qc = QuantumCircuit(4)
qc.h(0)
qc.x(1)
qc.cp(phi, 0, 1)
qc.h(2)
qc.x(3)
qc.p(disturb, 3)
qc.cp(phi, 2, 3)
qc.save_statevector()
job = sim.run(transpile(qc, sim), shots=256)
state = job.result().get_statevector()
plot_bloch_multivector(state)
```

```python
from qiskit import QuantumCircuit
import math
```

### Toffoli gate

$$ T|a>|b>|c> = |a>|b>|c \oplus (a \land b)> $$

```python
toffoli = QuantumCircuit(3)

toffoli.ccx(0,1,2)

toffoli.draw(output='mpl')
```

```python
# another way to realize toffoli gate

toffoli = QuantumCircuit(3)
```

### multi-controlled gate

```python
from qiskit.circuit.library import MCXGate

gate = MCXGate(4)
circuit = QuantumCircuit(5)
circuit.append(gate, [0, 1, 4, 2, 3])
circuit.draw('mpl')
```

### Rx gate and Controlled-Rx gate

$$ R_x(\theta) = \begin{bmatrix} \cos(\theta/2) & -i\sin(\theta/2) \\ -\sin(\theta/2) & \cos(\theta/2) \end{bmatrix} $$

```python
circuit = QuantumCircuit(2)
circuit.rx(math.pi/4,0)
circuit.crx(math.pi/6,0,1)
circuit.draw(output='mpl')
```

### Ry gate and Controlled-Ry gate
$$ R_y(\theta) = \begin{bmatrix} \cos(\theta/2) & -\sin(\theta/2) \\ \sin(\theta/2) & \cos(\theta/2) \end{bmatrix} $$

```python
circuit = QuantumCircuit(2)
circuit.ry(math.pi/4,0)
circuit.cry(math.pi/6,0,1)
circuit.draw(output='mpl')
```

### Rz gate and Controlled-Rz gate
$$ R_z(\theta) = \begin{bmatrix} 1 & 0 \\ 0 & e^{i\theta} \end{bmatrix} $$
### Qiskit sometimes uses cp instead of CRz

```python
circuit = QuantumCircuit(2)
circuit.rz(math.pi/4,0)
circuit.crz(math.pi/6,0,1)
circuit.draw(output='mpl')
```

```python
circuit = QuantumCircuit(2)
circuit.cp(math.pi/6,0,1)
circuit.draw(output='mpl')
```
